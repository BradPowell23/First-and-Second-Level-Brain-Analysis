```python
import nilearn
import numpy as np
import pandas as pd
from nilearn import plotting
from nilearn import image
from nilearn.glm.first_level import make_first_level_design_matrix
```

# Run 1

```python
#events files
events_view = pd.read_csv("/Users/bradpowell/Desktop/sub-01/ses-01/Events/view_first_run.tsv", sep='\t')
events_recall = pd.read_csv("/Users/bradpowell/Desktop/sub-01/ses-01/Events/recall_first_run.tsv", sep='\t')
```

```python
events_view['trial_type'] = 'view'
events_recall['trial_type'] = 'recall'
events = pd.concat([events_view, events_recall])
events
```

```python
timings = np.arange(0, 901 + 329)
```

```python
make_first_level_design_matrix(timings, events, drift_model=None)#.plot(subplots=True)
```
