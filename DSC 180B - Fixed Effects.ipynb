```python
from nilearn.glm.contrasts import compute_fixed_effects

contrast_imgs = ['dsc180_contrasts\sub-06_run-01_effect_size.nii', 
                 'dsc180_contrasts\sub-06_run-02_effect_size.nii', 
                 'dsc180_contrasts\sub-06_run-03_effect_size.nii']
variance_imgs = ['dsc180_contrasts\sub-06_run-01_effect_variance.nii', 
                 'dsc180_contrasts\sub-06_run-02_effect_variance.nii', 
                 'dsc180_contrasts\sub-06_run-03_effect_variance.nii']
```

```python
from nilearn import plotting
```

```python
fixed_fx_contrast, fixed_fx_variance, fixed_fx_stat = compute_fixed_effects(
    contrast_imgs, variance_imgs)
plotting.plot_stat_map(
    fixed_fx_stat, threshold=3.0)
```

```python
fixed_fx_contrast.to_filename('dsc180_contrasts\sub-06_run-all_effect_size')
fixed_fx_variance.to_filename('dsc180_contrasts\sub-06_run-all_effect_variance')
```
